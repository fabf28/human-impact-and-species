###Human Population 1970 - 2023

```python
df_pop
```

```python
all_pops = list(df_pop["Population"].apply(lambda x: int("".join(x.split(",")))))
all_pops
```

```python
the_pops = []

index = 0
for i in range(1990, 2017):
  pop = 0
  for j in range(4):
    pop += all_pops[index]
    index += 1
  the_pops.append(pop/4)
the_pops
```

```python
#checking
print(the_pops[0], the_pops[1], the_pops[2], the_pops[3])
print((all_pops[0] + all_pops[1] + all_pops[2] + all_pops[3])/4, (all_pops[4] + all_pops[5] + all_pops[6] + all_pops[7])/4)
```

```python
dict_pop = {
    "Year": list(range(1990, 2017)),
    "Population": the_pops
}
df_pop1 = pd.DataFrame(dict_pop)
df_pop1
```

###Combining All 1990 - 2018

```python
df1 = pd.merge(df_pop1, df_ghg, on = "Year")
df = pd.merge(df1, df_csi, on = "Year")
df
```

```python
#function for linear regression
def print_stats(slope, intercept, r, p, std_err):
  print("slope: ", slope)
  print("y intercept: ", intercept)
  print("r value: ", r)
  print("r squared value: ", r ** 2)
  print("p value: ", p)
  print("standard deviation error: ", std_err)
```

```python
#function for linear regression
def my_linear_regression(df, xval, yval, the_title, xlab, ylab):
  df.plot.scatter(x = xval, y = yval, title = the_title, xlabel = xlab, ylabel = ylab)

  x = df[xval]
  y = df[yval]

  slope, intercept, r, p, std_err = stats.linregress(x, y)

  def myfunc(x):
    return slope * x + intercept

  mymodel = list(map(myfunc, x))
  plt.plot(x, mymodel)
  plt.show()
  print_stats(slope, intercept, r, p, std_err)
  print("pearson correlation coefficient: ", "...")
```
